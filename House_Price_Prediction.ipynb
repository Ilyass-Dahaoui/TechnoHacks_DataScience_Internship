Linear Regression Model

```python
non_numeric_columns = data.select_dtypes(exclude=['number'])

# Afficher les colonnes non numÃ©riques
print("Colonnes non numÃ©riques :")
print(non_numeric_columns.columns)
```

```python
le = LabelEncoder()
data['date_encoded'] = le.fit_transform(data['date'])

# Supprimer la colonne 'date' originale
data.drop('date', axis=1, inplace=True)
```

```python
x=data.drop("price",axis=1)
y=data["price"]
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

Evaluate the model

```python
mse = mean_squared_error(y_test, y_pred)
rmse = mse**0.5
print("Root Mean Squared Error (RMSE):", rmse)
r_squared = r2_score(y_test, y_pred)
print(r_squared*100,"%")
```
